```python
# Assuming the DataFrame is already loaded as rides_cp

# Convert datetime columns
rides_cp['started_at'] = pd.to_datetime(rides_cp['started_at'])
rides_cp['ended_at'] = pd.to_datetime(rides_cp['ended_at'])

# Create duration column in minutes
rides_cp['duration'] = (rides_cp['ended_at'] - rides_cp['started_at'])

# Describe the duration column
duration_description = rides_cp['duration'].describe()
duration_description
```
